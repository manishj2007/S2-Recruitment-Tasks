# Task 2: Exploratory Data Analysis (EDA)

Goal:
To explore the Students Performance dataset using visualizations and identify patterns, relationships, and possible biases.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("StudentsPerformance.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
plt.figure()
sns.histplot(df["math score"])
plt.title("Distribution of Math Scores")
plt.xlabel("Math Score")
plt.ylabel("Number of Students")
plt.show()
```

# Distribution of Math Scores

The math scores appear to be roughly normally distributed, with most students scoring between 60 and 80.
There are fewer students with very low or very high scores, indicating that extreme values are less common.


```python
plt.figure()
sns.scatterplot(x=df["reading score"], y=df["writing score"])
plt.title("Reading Score vs Writing Score")
plt.xlabel("Reading Score")
plt.ylabel("Writing Score")
plt.show()
```

# Reading Score vs Writing Score

The points form a clear upward pattern, showing that students who score higher in reading also tend to score higher in writing.
This suggests that performance in these two subjects is closely linked.



```python
plt.figure()
sns.boxplot(x=df["test preparation course"], y=df["reading score"])
plt.title("Reading Scores by Test Preparation Course")
plt.xlabel("Test Preparation Course")
plt.ylabel("Reading Score")
plt.show()
```

# Effect of Test Preparation Course on Reading Scores

Students who completed the test preparation course generally
show higher reading scores compared to those who did not.
The difference in medians suggests that test preparation may
had a positive impact on reading performance.


```python
plt.figure()
sns.boxplot(x=df["lunch"], y=df["math score"])
plt.title("Math Scores by Lunch Type")
plt.xlabel("Lunch Type")
plt.ylabel("Math Score")
plt.show()
```

# Math Scores by Lunch Type

Students receiving standard lunch seem to have slightly higher math scores compared to those receiving free or reduced lunch.
This suggests that external factors, such as access to resources,can influence academic performance.


```python
average_scores = df[["math score", "reading score", "writing score"]].mean()
```

```python
plt.figure()
average_scores.plot(kind="bar")
plt.title("Average Scores Across Subjects")
plt.xlabel("Subject")
plt.ylabel("Average Score")
plt.show()
```

# Average Scores Across Subjects
The bar chart shows that average reading and writing scores are slightly higher than math scores.
This indicates that students generally perform better in language-based subjects compared to mathematics.
